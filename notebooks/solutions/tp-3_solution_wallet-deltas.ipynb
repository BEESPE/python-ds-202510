Charger les 2 fichiers CSV inventaires dans des dataframes pandas.

```python
import os
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
```

```python
DATA_DIR = Path("../../data")
INPUT_DIR = Path(DATA_DIR, "portfolio-inventories")
OUTPUT_DIR = Path(DATA_DIR, "output")

if not os.path.exists(Path(OUTPUT_DIR)):
    os.makedirs(Path(OUTPUT_DIR))
```

```python
PORTFOLIO_CODE = "Portoflio Code"
ASSET_CODE = "Asset Code"
QUANTITY = "Quantity"
```

```python
portfolio_1_df = pd.read_csv(
    Path(INPUT_DIR, "day1.ptf_inventories.csv"),
    sep=";",
)

portfolio_2_df = pd.read_csv(
    Path(INPUT_DIR, "day2.ptf_inventories.csv"),
    sep=";",
)
```

Explorer ces fichiers pour essayer d'en comprendre la structure.

Déterminer s'il y a des lignes en "doublon".

```python
portfolio_1_df[
    portfolio_1_df.duplicated(subset=[PORTFOLIO_CODE, ASSET_CODE], keep=False)
]
```

Agréger les lignes multiples associées à un même couple (portfolio, asset), en utilisant une fonction d'agrégation pertinente.

```python
aggregated_portfolio_1_df = portfolio_1_df[
    [PORTFOLIO_CODE, ASSET_CODE, QUANTITY]
].groupby(
    [PORTFOLIO_CODE, ASSET_CODE],
    as_index=False,
).sum(QUANTITY)
aggregated_portfolio_1_df
```

```python
aggregated_portfolio_2_df = portfolio_2_df[
    [PORTFOLIO_CODE, ASSET_CODE, QUANTITY]
].groupby(
    [PORTFOLIO_CODE, ASSET_CODE],
    as_index=False,
).sum(QUANTITY)
aggregated_portfolio_2_df
```

Ecrire un petit script permettant de calculer les variations de chaque asset de chaque portfolio, dès lors qu'il y a eu une variation non nulle. Stocker les résultats dans un dataframe.

```python
merged_quantity_df = aggregated_portfolio_1_df.merge(
    aggregated_portfolio_2_df,
    how="outer",
    on=[PORTFOLIO_CODE, ASSET_CODE],
    suffixes=["_before", "_after"],
).fillna(0)
```

```python
merged_quantity_df
```

```python
merged_quantity_df["variation"] = merged_quantity_df[f"{QUANTITY}_after"] - merged_quantity_df[f"{QUANTITY}_before"]
```

```python
non_zero_variation_df = merged_quantity_df[
    merged_quantity_df["variation"] != 0
]
```

```python
len(non_zero_variation_df)
```

Exporter le dataframe des variations au format CSV.

```python
non_zero_variation_df.to_csv(Path(OUTPUT_DIR, "non_zero_variation.csv"))
```

Visualiser la distribution des variations non nulles.

```python
plt.figure()
plt.hist(x=non_zero_variation_df["variation"])
plt.xlabel("Variations (regroupées en classes)")
plt.ylabel("Nombre d'occurences")
plt.title("Distribution des variations des compositions des portefeuilles")
plt.show()
```

Visualiser la composition d'un portefeuille à une date donnée.

```python
portfolio_number = 111111
```

```python
# pandas
pie_mask = (
    (aggregated_portfolio_1_df[PORTFOLIO_CODE] == 111111)
    &
    (aggregated_portfolio_1_df[QUANTITY] > 0)
)
pie_df = aggregated_portfolio_1_df[pie_mask].sort_values(QUANTITY, ascending=False)
```

```python
plt.figure()
plt.pie(
    x=pie_df[QUANTITY],
    labels=pie_df[ASSET_CODE],
)
plt.title(f"Composition du portefeuille {portfolio_number}")
plt.show()
```

Regrouper les assets représentant moins de 2% du total dans une catégorie "others".

```python
threshold = 0.02 * pie_df[QUANTITY].sum()
other_mask = pie_df[QUANTITY] < threshold

major_assets = pie_df[~other_mask]
minor_assets = pie_df[other_mask]
```

```python
major_assets
```

```python
label_list = major_assets[ASSET_CODE].tolist()
value_list = major_assets[QUANTITY].tolist()

if not minor_assets.empty:
    label_list.append("others")
    value_list.append(minor_assets[QUANTITY].sum())
```

```python
label_list
```

```python
value_list
```

```python
plt.figure()
plt.pie(
    x=value_list,
    labels=label_list,
)
plt.title(f"Composition du portefeuille {portfolio_number}")
plt.show()
```

```python
plt.figure()
plt.pie(
    x=value_list,
    labels=label_list,
    wedgeprops={
        "edgecolor": "white",
        "linewidth": 1.2,
    },
    autopct="%1.1f%%",
)
plt.title(f"Composition du portefeuille {portfolio_number}")
plt.tight_layout()
plt.savefig(Path(OUTPUT_DIR, f"portfolio_{portfolio_number}.png"), dpi=200)
plt.show()
```
