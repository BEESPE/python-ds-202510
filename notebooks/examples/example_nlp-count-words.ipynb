Comptage de mots dans un fichier texte (moby_dick.txt).

```python
from collections import Counter
import re
```

```python
with open('../../data/moby_dick.txt', 'r', encoding='utf-8') as file:
    text = file.read().lower()
```

```python
words = re.findall(r'\b\w+\b', text)
word_counts = Counter(words)
```

```python
for word, count in word_counts.most_common():
    print(f'{word}: {count}')
```
